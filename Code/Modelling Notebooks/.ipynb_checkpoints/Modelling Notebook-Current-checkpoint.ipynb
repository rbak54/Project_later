## C -contact duration

$s=\frac{c_u-m_T}{1.96} \\
N_D=  \frac{1}{\sqrt{2\pi}\times s} \times e^\frac{-(T-m_T)^2}{2 s^2}\\ 
d(T)= N_D \times m_d \times \sqrt{2\pi} \times s \\
C= \omicron \times d \\\\
$  
s is from $CI=\mu \pm z s$, then rearrange. 
N_D is just the normal distribution.
d(T) is the normal distribution scaled such that the maximum duration of contact is the maximum value. This works because when $\mu$=$x$ then $N_D=\frac{1}{\sqrt{2\pi}s}$ 


$m_T$ and $m_d$ are the temperature and duration of contact at the maximum duration. $c_l$ and $c_u$ is the interval at which 95\%  of the duration is likely to occur. The standard deviation $s$ is the calculated such that 95\% of the area falls between $c_l$ and $c_u$.  $N_D$ is the normal distribution with standard distribution $s$ and mean $m_T$. $T$ is the temperature. $d$ is the duration of contact and is multiplied by a scaling factor to stretch the function in the y direction such that the maximum temperacture corresponds to the maximum duration. $C$ is the contact duration multiplied by scaling factor $\omicron$. 

```r
duration_function <- function(m , c, T_range){
    T = seq(T_range[1],T_range[2],length.out = 10000)
    c_u = c[2]
    m_T = m[1]
    m_d = m[2]
    s = (c_u- m_T) / 1.96
    N_D =  (1 / (sqrt(2 * pi) * s)) *  exp( ( - ( T - m_T ) ^2 ) / (2 * s^2))
    d = N_D * m_d * sqrt(2 * pi) * s 
    #Multiply by a factor such that the mean of D is the mean contact time
    #d=N_D*19/mean(N_D)

    #plot(T,N_D,type="s")
    plot(T,d,type="s")
}
```

## S - survivability of virus


Equation for exponential virus decay:  
$V=V_0 \times e^{qt} $   
where V_0 is the initial virus density and q is the rate of decay.   
q can be calculated from substituting V_0 and the coordinates at another time into the equation. 
(alternatively:$V= q^t \times V_0$)
This would be more accurate if you included that q will vary with time, but here just getting rough idea to understand.  
$S= f(T) \times l$  
    where f(T) is bassed on a  quadratic equation made from fitting a quadratic equation to  the decay rate at different temperatures

```r
findq<-function(V_0,timepoint){
    x<-seq(0,10,length.out = 10000)
    q<-log(timepoint[2]/V_0)/timepoint[1]
    V<-V_0 * exp(q*x)
  #  plot(x,V,type="s")
    return(q)
    }
```

q(T) is a quadratic equation fit to the growth rate at different temperatures using the vertex and one other point and temperature limits. As this results in negative numbers, it needs to be scaled to maintain the shape of the curve while removing negative values. Current plan- f(T)= q(T) - min(q) (then multiply f(x) by a scaling factor)

```r
make_quadratic_formal <- function(x,y, T_range){
    T=seq(T_range[1],T_range[2],length.out = 10000)
    f = lm(y ~ poly(x, degree=2))
    f_T <- predict.lm(f,data.frame(x=T))
    plot(T,f_T,"s",col="blue")
    points(x,y)
}

x<-c(4,22,37,56,70)
x<-x[1:4]
y<-c(findq(6.51,c((14*24*60-30)/(24*60),6.04)),findq(6.51,c((7*3*60)/(24*60),3.48)),findq(6.57,c((1*3*60-30)/(24*60),3.23)),findq(6.65,c(10/(24*60),3.84)),findq(5.34,c(5/(24*60),0)))
y<-y[1:4]
T_range=c(0,60)


make_quadratic_formal(x,y,T_range=c(-10,60))
```

```r
make_quadratic_positive <- function(x,y, T_range){
    T=seq(T_range[1],T_range[2],length.out = 10000)
    f = lm(y ~ poly(x, degree=2))
    f_T <- predict.lm(f,data.frame(x=T))
    min <--min(f_T)
    f_T <- predict.lm(f,data.frame(x=T)) + min
    plot(T,f_T,"s",col="blue")
    points(x,y+min)
}

x<-c(4,22,37,56,70)
x<-x[1:4]
y<-c(findq(6.51,c((14*24*60-30)/(24*60),6.04)),findq(6.51,c((7*3*60)/(24*60),3.48)),findq(6.57,c((1*3*60-30)/(24*60),3.23)),findq(6.65,c(10/(24*60),3.84)),findq(5.34,c(5/(24*60),0)))
y<-y[1:4]
T_range=c(0,60)


make_quadratic_positive(x,y,T_range=c(-10,60))
```

## Quadratic for C

Current plan- quadratic equation, with coordinates of the maximum point and one other point. Most likely the contact at min or max temperature
$C= \omicron \times d \\
d=- \frac{ m_d - p_d}{(p_T-m_T)^2 } \times (T-m_T)^2 + m_d \\
$  
where d is the duration of contact, $\omicron$ is a constant and $T$ is the temperature. $m_T$ and $m_d$ are the temperature and duration of contact at the maximum duration. $p_T$ and $p_d$ are the temperature and duration at another point. 
duration_function <- function(m , p , srange){
    T = seq(srange[1], srange[2] , length.out = 1000)
    d= -((m[2] - p[2])/(p[1]-m[1])^2)*(T-m[1])^2 + m[2]
    plot(T,d,type="s")
}
duration_function(m = c(22,38),p= c(-10,20),c(-10,40))

```r
\beta = \beta_0  \times C(T) \times S(T) \times H  \\
H=(1 - m\times \frac{I^3} {k^3 + I^3}) \\
d(T)= N(m_T,s^2) \times m_d \times \sqrt{2\pi s} \\
C(T)= \omicron \times d \\
q(T)=aT^2 + bT + c\\
S(T)=l \times q(T) \\
  \frac{dS}{dt} = \nu  N - \frac{\beta(T)  I  S} {N} - \mu S\\
  \frac{dE}{dt}=  \frac{\beta(T) I  S} {N} - (\sigma + \mu) E\\
  \frac{dI}{dt} = \sigma  E - (\alpha + \mu + \gamma) I\\
  \frac{dR}{dt} = \gamma I - \mu R  
```

```r
# negatives not appropriate for beta-so convert by - the minimum value. not elegent! but a slolutionb
make_quadratic_positive <- function(vertex , point , T_range){
   #vertex is the maxima or minima, point is another point on the curve
        T = seq(T_range[1], T_range[2] , length.out = 1000)
        q_T =  ( ( point[2] - vertex[2]) / (point[1] - vertex[1]) ^2) * (T-vertex[1])^2 + vertex[2] 
        
    #for parameters of quadratic equation
      a =( ( point[2] - vertex[2]) / (point[1] - vertex[1]) ^2) 
      b = ( ( point[2] - vertex[2]) / (point[1] - vertex[1]) ^2)  * -2 * vertex[1]   
      c = ( ( point[2] - vertex[2]) / (point[1] - vertex[1]) ^2)  * vertex[1]^2 + vertex[2] - min(q_T) 
      q_T = a * T^2 + b * T + c
      plot(T,q_T,type="s")
    return(c(a,b,c))
}


x<-c(4,22,37,56,70)
#until q undetectable for all
y<-c(findq(6.51,c((14*24*60-30)/(24*60),6.04)),findq(6.51,c((7*3*60)/(24*60),3.48)),findq(6.57,c((1*3*60-30)/(24*60),3.23)),findq(6.65,c(10/(24*60),3.84)),findq(5.34,c(5/(24*60),0)))
make_quadratic_positive(vertex= c(x[1],y[1]),point= c(x[4],y[4]),T_range=c(-10,60))


```

```r
make_quadratic <- function(vertex , point , T_range){
   #vertex is the maxima or minima, point is another point on the curve
        T = seq(T_range[1], T_range[2] , length.out = 1000)
        M =  ( ( point[2] - vertex[2]) / (point[1] - vertex[1]) ^2) * (T-vertex[1])^2 + vertex[2]
       #for parameters of quadratic equation
        a = ( ( point[2] - vertex[2]) / (point[1] - vertex[1]) ^2)
        b = ( ( point[2] - vertex[2]) / (point[1] - vertex[1]) ^2) * -2 * vertex[1]  
        c = ( ( point[2] - vertex[2]) / (point[1] - vertex[1]) ^2) * vertex[1]^2 + vertex[2]
    plot(T,M,type="s")
    return(c(a,b,c))
}

#from above
x<-c(4,22,37,56,70)

y<-c(findq(6.51,c((14*24*60-30)/(24*60),6.04)),findq(6.51,c((7*3*60)/(24*60),3.48)),findq(6.57,c((1*3*60-30)/(24*60),3.23)),findq(6.65,c(10/(24*60),3.84)),findq(5.34,c(5/(24*60),0)))
make_quadratic(vertex= c(x[1],y[1]),point= c(x[4],y[4]),T_range=c(-10,60))

```
